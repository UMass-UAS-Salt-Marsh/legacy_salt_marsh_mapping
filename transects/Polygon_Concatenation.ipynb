```python
import os
import geopandas as gpd
import pandas as pd
```

```python
# Concatenate mapping units

# Specify which folder to look in to concatenate shapefiles
workingFolder = input("Enter file path of working folder for specified site containing shapefiles to concatenate --")
workingFolder = workingFolder[1:-1]
file = os.listdir(workingFolder)

# Look for all files ending in ".shp"
path = [os.path.join(workingFolder, i) for i in file if ".shp" in i]

# Concatenate to file specified in last line of code
gdf = gpd.GeoDataFrame(pd.concat([gpd.read_file(i) for i in path],
                        ignore_index=True), crs=gpd.read_file(path[0]).crs)

# Write new concatenated shapefile to final folder
finalOut = input("Enter file path of final transect folder to write final data product to -- ")
# Edit string appended to finalOut to change name of final output file
finalOut = finalOut[1:-1] + "\\" + input('Enter name of polygons ') + ".shp"

# Delete unneeded attributes for training data
# del gdf['Pre Angle']
# del gdf['Post Angle']
# del gdf['mu']
# del gdf['Class mu']
# del gdf['Pre/Post']

# Write out file
gdf.to_file(finalOut)
```
